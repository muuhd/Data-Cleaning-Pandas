```python
import pandas as pd
```

```python
#checking for encoding/delimiter error and specifying dtype for string/bool columns

df = pd.read_csv('Messy_Employee_dataset.csv',
                 encoding = 'utf-8',
                 sep = ',',
                dtype = {'Employee_ID':'string',
                         'First_Name':'string',
                         'Last_Name':'string',
                         'Department_Region':'string',
                         'Status':'string',
                         'Email':'string',
                         'Performance_Score':'string',
                         'Remote_Work':'bool'
                         }
                )
df 
```

```python
#fixing column types for numerical columns + date column

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors='coerce')

df
```

```python
#checking for missing

df.isnull().sum()
```

```python
#standardising whitespace and capitalisation
df.columns = df.columns.str.lower().str.replace(' ', '_').str.strip()
df

#setting employee_id as index
df.set_index('employee_id', inplace=True)
```

```python
#fixing phone column
df['phone'] = df['phone'].astype(str)
df
```

```python
#handle missing data

df.isnull().sum()

df.dropna(subset = ['salary'], inplace=True) #cannot really impute any value for salary as it is specific to job role and other quantifiers - best to be removed

df['age'] = df['age'].fillna(df['age'].median())
```

```python
#check again for missing values
df.isnull().sum()
```

```python
#remove '-' sign for phone numbers
df['phone'] = df['phone'].str.replace(r'\D','', regex=True)

df
```

```python
#export to .csv
df.to_csv('cleaned_employee_dataset.csv', index = True)
```
